### Jumlah buku yang dirilis setelah tanggal 1 Januari 2000

```python
# menghitung jumlah buku
query = '''
         SELECT COUNT(*) AS book_count
         FROM books
         WHERE publication_date > '01-01-2000'
        '''

data = pd.io.sql.read_sql(query, con = engine)
data
```

Jumlah buku yang dirilis setelah tanggal 1 Januari 2000 sebanyak 819 buku.

### Jumlah ulasan pengguna dan rating rata-rata untuk setiap buku

```python
# menghitung jumlah ulasan pengguna dan rating rata-rata untuk setiap buku
query = '''
        SELECT 
            b.title, 
            COUNT(r.rating_id) AS total_reviews, 
            AVG(r.rating) AS avg_rating
        FROM books b
        LEFT JOIN 
            ratings r ON b.book_id = r.book_id
        GROUP BY 
            b.book_id, 
            b.title
        ORDER BY 
            total_reviews DESC;
        '''

data = pd.io.sql.read_sql(query, con = engine)
data
```

Dari tabel di atas dapat diketahui bahwa jumlah ulasan pengguna tertinggi yaitu buku Twilight (Twilight#1) dengan total jumlah ulasan sebanyak 160. Sedangkan jumlah ulasan terendah yaitu buku Disney's Beauty and the Beast yaitu hanya 1 pengguna yang mengulas buku tersebut.

### Penerbit yang telah menerbitkan jumlah buku terbanyak dengan jumlah halaman lebih dari 50

```python
# melihat penerbit yang telah menerbitkan jumlah buku terbanyak dengan jumlah halaman lebih dari 50
query = '''
        SELECT 
            p.publisher,
            COUNT(b.book_id) AS books
        FROM publishers p
        JOIN
            books b ON p.publisher_id = b.publisher_id
        WHERE
            b.num_pages > 50
        GROUP BY
            p.publisher
        ORDER BY
            books DESC
        LIMIT 5;    
        '''

data = pd.io.sql.read_sql(query, con = engine)
data
```

Tabel di atas memuat 5 penerbit yang telah menerbitkan buku terbanyak dengan jumlah halaman lebih dari 50 lembar. Dapat diketahui bahwa penerbit `Penguin Books` menjadi penerbit yang paling banyak menerbitkan buku.

### Penulis dengan rating rata-rata buku tertinggi, dengan mempertimbangkan buku yang memiliki rating minimal 50

```python
# melihat penulis dengan rating rata-rata buku tertinggi, dengan mempertimbangkan buku yang memiliki rating minimal 50
query = '''
        SELECT 
            a.author, 
            AVG (avg_rating) AS avg_rating
        FROM authors a
        JOIN 
            books b ON a.author_id = b.author_id
        JOIN
        (
            SELECT 
                bq.book_id,
                bq.title,
                COUNT (rq.rating_id) AS rating_count,
                AVG (rq.rating) AS avg_rating
            FROM books bq
            JOIN
                ratings rq ON rq.book_id = bq.book_id
            GROUP BY
                bq.book_id,
                bq.title
                
        ) AS subq ON subq.book_id = b.book_id
        WHERE 
            rating_count >= 50
        GROUP BY 
            a.author_id, 
            a.author
        ORDER BY 
            avg_rating DESC
        LIMIT 5;
        '''

data = pd.io.sql.read_sql(query, con = engine)
data
```

Tabel di atas memuat 5 penulis dengan rating rata-rata tertinggi dengan kriteria jumlah minimal ulasan sebanyak 50. Penulis `J.K. Rowling/Mary GrandPré` menjadi penulis dengan rating rata-rata tertinggi dengan rata-rata 4.28.

### Jumlah rata-rata teks ulasan di antara pengguna yang memberi rating terhadap lebih dari 50 buku

```python
# menghitung jumlah rata-rata teks ulasan di antara pengguna yang memberi rating terhadap lebih dari 50 buku
query = '''
        SELECT 
            AVG (review_count) AS avg_review_lenght
        FROM 
        (
            SELECT 
                r.username,
                COUNT (DISTINCT re.review_id) AS review_count
            FROM
                ratings r
            LEFT JOIN
                reviews re ON r.username = re.username
            WHERE
                r.username IS NOT NULL AND r.book_id IS NOT NULL
            GROUP BY
                r.username
            HAVING COUNT(DISTINCT r.book_id) > 50
        ) AS user_review
        
        '''

data = pd.io.sql.read_sql(query, con = engine)
data
```

Rata-rata jumlah teks ulasan di antara pengguna yang telah mengulas lebih dari 50 buku adalah 24.
